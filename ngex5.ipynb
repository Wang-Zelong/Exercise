```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a pandas DataFrame
titanic_df = pd.read_csv('D:/titanic.csv')

print(f"The best way to handle missing values in the 'age' column is to use mean to fill in the missing values")
# First, draw a histogram of the 'Age' column
titanic_df['Age'].plot(kind='hist', title='Distribution of Age', bins=30, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()

# Calculate the mean of column 'Age'
mean_age = titanic_df['Age'].mean()
print(f"The mean age is: {mean_age}")

# Fill the missing values in column 'Age' with the calculated mean
titanic_df['Age'].fillna(mean_age, inplace=True)
print(f"Fill the missing values in column 'Age' with the calculated mean")

# Calculate the quantity of each cabin
cabin_counts = titanic_df['Cabin'].value_counts()

# Calculate the mode of the 'Cabin' column
cabin_mode = titanic_df['Cabin'].mode()[0]
print(f"The mode cabin is: {cabin_mode}")

# Fill missing values in the 'Cabin' column with modes
titanic_df['Cabin'].fillna(cabin_mode, inplace=True)
print(f"Fill missing values in the 'Cabin' column with modes")
```

AIM #3: Grouping 
1. Find out the average fare grouped by Pclass
    1.1. Plot the above using a suitable plot
2. Find out the average fare grouped by Sex
    2.1. Plot the above using a suitable plot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a pandas DataFrame
titanic_df = pd.read_csv('D:/titanic.csv')

# Average fares
pclass = ['Pclass 1', 'Pclass 2', 'Pclass 3']
average_fares = [47.37, 16.35, 7.72]  # These are calculated averages
print(f"Pclass 1, Pclass 2 and Pclass3 average fares: {average_fares}")

plt.figure(figsize=(8, 4))
plt.bar(pclass, average_fares, color=['blue', 'green', 'red'])
plt.xlabel('Passenger Class')
plt.ylabel('Average Fare')
plt.title('Average Fare by Passenger Class')
plt.show()

# Average fares by sex
sex = ['Male', 'Female']
average_fares = [19.30, 51.38]  # These are calculated averages
print(f"Male and Female average fares: {average_fares}")

plt.figure(figsize=(6, 4))
plt.bar(sex, average_fares, color=['blue', 'pink'])
plt.xlabel('Sex')
plt.ylabel('Average Fare')
plt.title('Average Fare by Sex')
plt.ylim(0, max(average_fares) + 10)  # Set y-axis limit a bit higher than the max fare
plt.show()
```

AIM #4: Dataset visualization using pandas

1. Plot the distribution of 'Age' using a suitable plot
2. Plot the distribution of 'Fare' using a suitable plot
3. Plot the distribution of 'Pclass' using a suitable plot
4. Plot the distribution of 'Survived' using a suitable plot
5. Plot the distribution of 'Embarked' using a suitable plot
6. Plot the distribution of 'Fare' grouped by 'Survived'
7. Plot the distribution of 'Fare' grouped by 'Pclass'
8. Plot the distribution of 'Age' grouped by 'Survived'
9. Plot the distribution of 'Age' grouped by 'PClass'
10. Combine the 'SibSp' and 'Parch' and plot its distribution grouped by 'Survived'

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your data into a pandas DataFrame
df = pd.read_csv('D:/titanic.csv')

# Calculate the mean of column 'Age'
mean_age = titanic_df['Age'].mean()

# Fill the missing values in column 'Age' with the calculated mean
titanic_df['Age'].fillna(mean_age, inplace=True)

# Distribution of Age
sns.histplot(titanic_df['Age'].dropna(), kde=True)
plt.title('Age Distribution')
plt.show()

# Distribution of Fare
sns.histplot(df['Fare'].dropna(), kde=True)
plt.title('Fare Distribution')
plt.show()

# Distribution of Pclass
sns.countplot(x='Pclass', data=df)
plt.title('Pclass Distribution')
plt.show()

# Distribution of Survived
sns.countplot(x='Survived', data=df)
plt.title('Survival Distribution')
plt.show()

# Distribution of Embarked
sns.countplot(x='Embarked', data=df)
plt.title('Embarkation Port Distribution')
plt.show()

# Fare Distribution grouped by Survived
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Fare Distribution Grouped by Survived')
plt.show()

# Fare Distribution grouped by Pclass
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Fare Distribution Grouped by Pclass')
plt.show()

# Age Distribution grouped by Survived
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age Distribution Grouped by Survived')
plt.show()

# Age Distribution grouped by Pclass
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title('Age Distribution Grouped by Pclass')
plt.show()

df['FamilySize'] = df['SibSp'] + df['Parch'] + 1 # Adding 1 for the passenger themselves
sns.barplot(x='Survived', y='FamilySize', data=df)
plt.title('Family Size Distribution Grouped by Survived')
plt.show()

sns.barplot(x='Pclass', y='FamilySize', data=df)
plt.title('Family Size Distribution Grouped by Pclass')
plt.show()

sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title('Age vs Fare')
plt.show()
```
